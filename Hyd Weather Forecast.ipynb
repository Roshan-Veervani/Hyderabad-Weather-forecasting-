```python
df=df.dropna()
```

```python
df_max=df.rename(columns={'Date':'ds','MaxT':'y'})
df_max
```

```python
df_new=df_max[['ds','y']]
df_new
```

```python
df_new.info()
```

```python
df_new['y']=df_new['y'].astype(int)
```

```python
df_new.info()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(25,5))
plt.scatter(df_new['ds'],df_new['y'])
```

```python
import prophet
op=prophet.Prophet()
op.fit(df_new)
```

```python
forecast=op.make_future_dataframe(periods=150,freq='M')
forecast=op.predict(forecast)
forecast
```

```python
op.plot(forecast,xlabel = 'Date',ylabel = 'Max Temperature')
```

```python
op.plot_components(forecast)
```
